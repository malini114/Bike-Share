```python
import pandas as pd
import numpy as np
```

```python
q_2019 = pd.read_csv('bike-share-2019Q1.csv')
q_2020 = pd.read_csv('bike-share-2020Q1.csv')
```

```python
q_2019.info()
```

```python
q_2020.info()
```

```python
q_2019 = q_2019.rename(columns={
'trip_id': 'ride_id',
'bikeid': 'rideable_type',
'start_time': 'started_at',
'end_time': 'ended_at',
'from_station_name': 'start_station_name',
'from_station_id': 'start_station_id',
'to_station_name': 'end_station_name',

'to_station_id': 'end_station_id',
'usertype': 'member_casual'
})
```

```python
q_2019.info()
q_2020.info()
```

```python
# converting datatypes for concating
q_2019['ride_id'] = q_2019['ride_id'].astype(str)
q_2019['rideable_type'] = q_2019['rideable_type'].astype(str)
```

```python
#concat two datasets
all_trips = pd.concat([q_2019,q_2020],ignore_index=True)
```

```python
all_trips = all_trips.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng', 'birthyear', 'gender','tripduration'], errors='ignore')
```

```python
#value count in member_casual column
all_trips['member_casual'].value_counts()
```

```python
all_trips['member_casual'] = all_trips['member_casual'].replace({
    'Subscriber': 'member',
    'Customer': 'casual'
    })
```

```python
#check for reassign
all_trips['member_casual'].value_counts()
```

```python
#convert started_at and ended_at as datetime

all_trips['started_at'] = pd.to_datetime(all_trips['started_at'], errors='coerce')
all_trips['ended_at']   = pd.to_datetime(all_trips['ended_at'], errors='coerce')
```

```python
#add separate columns for date month and day

all_trips['date'] = all_trips['started_at'].dt.date
all_trips['day'] = all_trips['started_at'].dt.day
all_trips['month'] = all_trips['started_at'].dt.month
all_trips['year'] = all_trips['started_at'].dt.year
all_trips['day_of_week'] = all_trips['started_at'].dt.day_name()
```

```python
#ride_length calculation

all_trips['ride_length'] = (all_trips['ended_at']-all_trips['started_at']).dt.total_seconds()
```

```python
# Inspect the structure of the columns
print("\nData types after adding date columns and ride_length:\n")
all_trips.info()
```

```python
all_trips['ride_length'].count()
```

```python
all_trips_v2 = all_trips[(all_trips['start_station_name'] != "HQ QR") & 
                         (all_trips['ride_length'] >=0)].copy()
```

```python
all_trips_v2['ride_length'].count()
```

```python
#descriptive analysis on ride_length
all_trips_v2['ride_length'].describe()
```

```python
#compare member and casual users
print("\nRide length statistics grouped by member_casual:\n",
all_trips_v2.groupby('member_casual')['ride_length'].agg(['mean', 'median', 'max', 'min']))
```

```python
# See the average ride time by each day for members vs casual users
print("\nAverage ride length by member type and day of the week:\n",
all_trips_v2.groupby(['member_casual', 'day_of_week'])['ride_length'].mean())
```

```python
# ordering the days
days_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
all_trips_v2['day_of_week'] = pd.Categorical(all_trips_v2['day_of_week'],
categories=days_order, ordered=True)

all_trips_v2.groupby(['member_casual','day_of_week'])['ride_length'].mean()
```

```python
summary_stats = all_trips_v2.groupby(['member_casual', 'day_of_week']).agg(
number_of_rides=('ride_id', 'count'),
average_duration=('ride_length', 'mean')
).reset_index()
```

```python
print("\nSummary of rides and duration by rider type and weekday:\n", summary_stats)
```

```python
# Create a .csv file that you will visualize elsewhere
counts = all_trips_v2.groupby(['member_casual',
'day_of_week'])['ride_length'].mean().reset_index()
counts.to_csv('avg_ride_length.csv', index=False)
```
